```python
import pandas as pd
import matplotlib.pyplot as plt
import requests
import json
```

```python
response_API = requests.get('https://data.cityofnewyork.us/resource/uip8-fykc.json')
data = response_API.text
df = pd.DataFrame(eval(data))
```

```python
genders = []
all_genders = list(df['perp_sex'])
genders = list(set(all_genders))
arrests_in_each_gender = [0 for i in range(len(genders))]
for gender in all_genders:
    arrests_in_each_gender[genders.index(gender)] += 1
```

```python
df['perp_sex'].value_counts()
```

```python
fig = plt.figure(figsize = (10, 5)) 
# creating the bar plot
plt.bar(genders,arrests_in_each_gender)
plt.xlabel("Gender")
plt.ylabel("Arrests")
plt.show()
```

```python
age_groups = []
all_age_groups = list(df['age_group'])
age_groups = list(set(all_age_groups))
arrests_in_each_age_group = [0 for i in range(len(age_groups))]
for group in all_age_groups:
    arrests_in_each_age_group[age_groups.index(group)] += 1
```

```python
df['age_group'].value_counts()
```

```python
fig = plt.figure(figsize =(15, 10))
plt.pie(arrests_in_each_age_group, labels = age_groups, autopct='%1.1f%%',  explode=[0,0, 0, 0, 0.05], shadow=True)
plt.title('Percentage of Arrests in each Age Group')
plt.show()
```
